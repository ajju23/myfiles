```python
# Suggested starting point for use of Spacy
# Spacy documentation can be found here: https://spacy.io


# download model with: python -m spacy download en_core_web_(sm|lg)

# import spacy and load model
import spacy
activated = spacy.prefer_gpu()
nlp = spacy.load('en_core_web_lg')

# add custom pipelines to merge entities and noun phrases
# merge tokens for entities - this will ensure that words that belong to a single entity will be processed as one phrase
merge_ents = nlp.create_pipe("merge_entities")
nlp.add_pipe(merge_ents)

# merge tokens for noun chunks - this will ensure that words that belong to a single noun phrase will be processed as one phrase
merge_nps = nlp.create_pipe("merge_noun_chunks")
nlp.add_pipe(merge_nps)
```
